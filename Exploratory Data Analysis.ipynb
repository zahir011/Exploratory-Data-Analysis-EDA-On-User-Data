# Exploratory Data Analysis

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Data set:
user_data = pd.read_csv('UserData (2).csv')
opportunity_data = pd.read_csv('Opportunity Wise Data.csv')
```

```python
# Display the first few rows of each dataset for a quick overview
user_data_head = user_data.head(10)
opportunity_data_head = opportunity_data.head(10)

user_data_head,opportunity_data_head
```

```python
# Display basic information about each data set:
user_data_info = user_data.info()
opportunity_data_info = opportunity_data.info()

user_data_info, opportunity_data_info
```

```python
# Get basic descriptive statistics for both datasets
user_data_describe = user_data.describe(include='all')
opportunity_data_describe = opportunity_data.describe(include='all')

user_data_describe,opportunity_data_describe
```

# Set up visualizations

```python
# Set up seaborn style for plots
sns.set(style="whitegrid")
```

```python
# Histogram for User Age Distribution (if 'Age' column exists in user data)
if 'Age' in user_data.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(user_data['Age'], bins=20, kde=True)
    plt.title('User Age Distribution')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()

# Bar chart for Opportunity Status distribution (if 'Status Description' column exists)
if 'Status Description' in opportunity_data.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=opportunity_data, x='Status Description', order=opportunity_data['Status Description'].value_counts().index)
    plt.title('Opportunity Status Distribution')
    plt.xlabel('Status Description')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Scatter plot for Reward Amount vs. Skill Points (if both columns exist in opportunity data)
if 'Reward Amount' in opportunity_data.columns and 'Skill Points Earned' in opportunity_data.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=opportunity_data, x='Reward Amount', y='Skill Points Earned')
    plt.title('Reward Amount vs. Skill Points Earned')
    plt.xlabel('Reward Amount')
    plt.ylabel('Skill Points Earned')
    plt.show()

# Visualization of Reward Amount Histogram
if 'Reward Amount' in opportunity_data.columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(opportunity_data['Reward Amount'], bins=20, kde=True, color="green")
    plt.title('Reward Amount Distribution')
    plt.xlabel('Reward Amount')
    plt.ylabel('Frequency')
    plt.show()
```
